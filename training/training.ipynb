```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainingArguments, Seq2SeqTrainer, AutoTokenizer
from datasets import load_dataset
import torch

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("ArielUW/jobtitles")
```

```python
print(ds["train"].column_names)
```

```python
fragment = ds["train"].select(range(5))
for example in fragment:
    print(example)
```

```python

model_name = "facebook/m2m100_1.2B"
tokenizer = AutoTokenizer.from_pretrained(model_name, src_lang="pl", tgt_lang="pl")
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

if torch.cuda.is_available():  # Check if GPU is available
    model = model.to("cuda")  # Move model to GPU
```

```python

def preprocess_function(examples):
    inputs = examples['source_sentence']
    targets = examples['target_sentence']
    model_inputs = tokenizer(inputs,
                             text_target=targets,
                             max_length=128,
                             truncation=True,
                             padding="max_length")
    if torch.cuda.is_available():  # Check if GPU is available
        for key in model_inputs:
            model_inputs[key] = torch.tensor(model_inputs[key]).to("cuda")  # Convert to tensor and move to GPU
    return model_inputs

# Apply to train split
tokenized_train = ds["train"].map(preprocess_function, batched=True)

# Apply to validation split
tokenized_validation = ds["validation"].map(preprocess_function, batched=True)

# Apply to evaluation split
tokenized_evaluation = ds["test"].map(preprocess_function, batched=True)
```

```python
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)
```

```python
import evaluate

metric = evaluate.load("sacrebleu")
```

```python
import numpy as np


def postprocess_text(preds, labels):
    preds = [pred.strip() for pred in preds]
    labels = [[label.strip()] for label in labels]

    return preds, labels


def compute_metrics(eval_preds):
    preds, labels = eval_preds
    if isinstance(preds, tuple):
        preds = preds[0]
    decoded_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)

    labels = np.where(labels != -100, labels, tokenizer.pad_token_id)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)

    decoded_preds, decoded_labels = postprocess_text(decoded_preds, decoded_labels)

    result = metric.compute(predictions=decoded_preds, references=decoded_labels)
    result = {"bleu": result["score"]}

    prediction_lens = [np.count_nonzero(pred != tokenizer.pad_token_id) for pred in preds]
    result["gen_len"] = np.mean(prediction_lens)
    result = {k: round(v, 4) for k, v in result.items()}
    return result
```

```python
model.config.dropout = 0.2
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/drive/MyDrive/ML_class/neutral_job_title_rephraser_pl",
    logging_dir="/content/drive/MyDrive/ML_class/logs",
    eval_strategy="epoch",
    logging_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=32,
    per_device_eval_batch_size=32,
    gradient_accumulation_steps=4,
    gradient_checkpointing=True,
    optim="adafactor",
    weight_decay=0.01,
    save_total_limit=3,
    num_train_epochs=7,
    predict_with_generate=True,
    fp16=True, #change to bf16=True for XPU
    push_to_hub=True,
    load_best_model_at_end=True,     # Load best model at the end based on metric
    metric_for_best_model='bleu',
    logging_steps=100
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_evaluation,
    processing_class=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

trainer.train()
```

```python
torch.cuda.empty_cache()
```

```python
eval_results = trainer.evaluate()
print(f"\nEvaluation Results: {eval_results}")
```

```python
def push_to_hub_with_config(trainer, tokenizer):
    """Push model and tokenizer to hub"""
    # Save tokenizer
    tokenizer.save_pretrained(trainer.args.output_dir)

    # Push everything to hub
    trainer.push_to_hub(
        commit_message="Add model, config, and tokenizer",
        blocking=True  # Wait until the push is complete
    )

push_to_hub_with_config(trainer, tokenizer)
```
